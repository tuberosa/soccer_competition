```python
#ダミー変数化していないもの、横軸離散値の場合に横軸変数ごとに縦軸の平均を取ってプロット
def plot_ave(a,b='y'):
    x=df[a].values
    y=df[b].values
    temp_x=[]
    y_ave=[]
    
    for i in range(len(df)):
        if x[i] not in temp_x:
            temp_x.append(x[i])
            array=np.array(df[df[a]==df[a][i]][b].values)
            y_ave.append(np.average(array))
        else:
            None
            
    plt.scatter(temp_x,y_ave)
    plt.xlabel(a)
    plt.ylabel('average  '+b)
    plt.title(a+' vs average  '+ b)
    plt.show()
```

```python
#平均ではなく中央値の方が良い？
def plot_median(a,b='y'):
    x=df[a].values
    y=df[b].values
    temp_x=[]
    y_med=[]
    
    for i in range(len(df)):
        if x[i] not in temp_x:
            temp_x.append(x[i])
            array=np.array(df[df[a]==df[a][i]][b].values)
            y_med.append(np.median(array))
        else:
            None
            
    plt.scatter(temp_x,y_med)
    plt.xlabel(a)
    plt.ylabel(b)
    plt.title(a+' vs median  '+ b)
    plt.show()
```

```python
def counting(a):
    temp=pd.DataFrame({a:df[a].values})
    temp2=temp.apply(pd.value_counts)
    plt.scatter(temp2.index,temp2.values)
    plt.title(a+'_count')
    plt.show()
```
