```python
#はじめにスタジアムデータの住所に対して、各スタジアムの緯度と経度を取得
#参考
#https://qiita.com/paulxll/items/7bc4a5b0529a8d784673
import requests
from bs4 import BeautifulSoup
import time
from tqdm import tqdm

#  address_lにlistの形で住所を入れてあげると、[緯度,経度]を返す。緯度経度ともfloatででてくる。
#  関数に住所を渡すとき、listで与えず一つずつfor文で呼び出すとうまくいかなかったため引数はlistにした。
def get_lat_lon_from_address(address_l):
    url = 'http://www.geocoding.jp/api/'
    latlons = []
    for address in tqdm(address_l):
        payload = {'q': address}
        r = requests.get(url, params=payload)
        ret = BeautifulSoup(r.content,'lxml')
        if ret.find('error'):
            raise ValueError(f"Invalid address submitted. {address}")
        else:
            lat = float(ret.find('lat').string)
            lon = float(ret.find('lng').string)
            latlons.append([lat,lon])
            time.sleep(10)
    return latlons
```
