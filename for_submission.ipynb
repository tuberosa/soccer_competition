```python
import pandas as pd
```

```python
#train_data
be=pd.read_csv('with_best_eleven.csv')
be['section_val']=be['J1_section_val'].values+be['J2_section_val'].values
be['home_popularity']=be['pop_home'].values-be['not_pop_home'].values
be['away_popularity']=be['pop_away'].values-be['not_pop_away'].values
```

```python
#説明変数の選択
train=be.loc[:,['y','distance','capa','best_eleven','rain','cloudy','indoor','fri','sat','sun','holiday','NHKs','section_val','away_popularity','home_popularity']]
```

```python
test=pd.read_csv('for_test.csv').drop('Unnamed: 0',axis=1)
```

```python
test
```

```python
#学習
X=train.drop('y',axis=1)
y=train['y'].values

X_test=test.loc[:,['distance', 'capa', 'best_eleven', 'rain', 'cloudy','indoor', 'fri', 'sat', 'sun', 'holiday', 'NHKs','section_val', 'away_popularity', 'home_popularity']]
```

```python
#線形回帰
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X,y)
```

```python
y_test_pred=[]
for i in range(len(X_test)):
        y_test_pred.append(min(lr.predict(X_test)[i],X_test['capa'].values[i]))
test['linear_regression']=y_test_pred        
```

```python
#XGBoost
import xgboost as xgb
mod = xgb.XGBRegressor()
mod.fit(X, y)
```

```python
#何の変数をdf_puから落とすかしていすれば予測を返す関数を定義
def predic(ls,model):
    #インスタンスの作成
    md=model()
    #データを作る
    if ls==0:
        temp_train=X
        temp_test=X_test
    else:
        temp_train=X.drop(ls,axis=1)
        temp_test=X_test.drop(ls,axis=1)
    #回帰
    md.fit(temp_train,y)
    #testについて予測値をリストで与える関数(predicting(model):testの予測)
    y_test_pred=[]
    for i in range(len(X_test)):
        y_test_pred.append(min(md.predict(temp_test)[i],X_test['capa'].values[i]))
    return y_test_pred    
```

```python
predic(['best_eleven','capa'],LinearRegression)
```

```python
import numpy as np
```

```python
#予測の平均とる
train_predic_array=np.array([predic(['best_eleven','capa'],LinearRegression),predic(['away_popularity','home_popularity'],LinearRegression),predic(['best_eleven','home_popularity'],LinearRegression),predic(['away_popularity','capa'],LinearRegression)])
average_predic=np.mean(train_predic_array,axis=0)
```

```python
test['ave_lin']=average_predic
```

```python
test
```

```python
#XGBoostで平均
#予測の平均とる
train_predic_array=np.array([predic(['best_eleven','capa'],xgb.XGBRegressor),predic(['away_popularity','home_popularity'],xgb.XGBRegressor),predic(['best_eleven','home_popularity'],xgb.XGBRegressor),predic(['away_popularity','capa'],xgb.XGBRegressor)])
average_predic=np.mean(train_predic_array,axis=0)
```

```python
test['ave_xg']=average_predic
```
