```python
#各試合に対してawayチームの移動距離を求める
a=0.0
b=0.0

test['distance']=0.0
for i in range(len(test)):
    for k in range(len(df_stad_)):
        if df_stad_['name'][k]==test['stadium'][i]:
            a=np.array(df_stad_['position'][k])
        if df_stad_['team_y'][k]==test['away'][i]:
            b=np.array(df_stad_['position'][k])
    test['distance'][i]=np.linalg.norm(a-b)
        
```

```python
df_stad_=df_stad_.drop('team_x',axis=1)
```

```python
#天気をダミー変数化する
for i in range(len(test)):
    if '雨' in test['weather'][i]:
        test['weather'][i]='rain'
    elif '雪' in test['weather'][i]:
        test['weather'][i]='rain'
    elif '曇' in test['weather'][i]:
        test['weather'][i]='cloudy'
    elif '屋内' in test['weather'][i]:
        test['weather'][i]='indoor'
    elif '晴' in test['weather'][i]:
        test['weather'][i]='sunny'

temp3=pd.get_dummies(test['weather'])
test['rain']=temp3['rain']
test['cloudy']=temp3['cloudy']
test['indoor']=temp3['indoor']
test['sunny']=temp3['sunny']
```

```python
#曜日のダミー変数化
test['fri']=0
test['sat']=0
test['sun']=0
test['holiday']=0
test['weekday']=0



for i in range(len(test)):
    if '金' in test['gameday'][i]:
        test['fri'][i]=1
    elif '土' in test['gameday'][i]:
        test['sat'][i]=1
    elif '日' in test['gameday'][i]:
        test['sun'][i]=1
    elif '祝' in test['gameday'][i]:
        test['holiday'][i]=1
    else:
        test['weekday'][i]=1
```

```python
#NHKs:NHK総合で放映されていたら1そうでなかったら0
test['NHKs']=0
for i in range(len(test)):
    if 'ＮＨＫ総合' in test['tv'][i]:
        test['NHKs'][i]=1
```

```python
stad_capa = dict(zip(df_stad_['name'].values, list(df_stad_['capa'].values)))
```

```python
test['capa']=test['stadium'].map(stad_capa)
```

```python
#スコア差
test['abs_score_dif']=abs(test['home_score'].values-test['away_score'].values)
```

```python
#popularity
df_sorted=df.sort_values('y',ascending=True)
df_sorted_p=df_sorted.tail(250)
df_sorted_n=df_sorted.head(250)

ph=(pd.DataFrame(data=df_sorted_p['home'])).apply(pd.value_counts)
ph_dict = ph.to_dict(orient='dict')

test['pop_home']=(test['home'].map(ph_dict['home'])).fillna(0)

#away_popularity
pa=(pd.DataFrame(data=df_sorted_p['away'])).apply(pd.value_counts)
pa_dict = pa.to_dict(orient='dict')

test['pop_away']=(test['away'].map(pa_dict['away'])).fillna(0)

#homeでの人気のなさ
nh=(pd.DataFrame(data=df_sorted_n['home'])).apply(pd.value_counts)
nh_dict = nh.to_dict(orient='dict')

test['not_pop_home']=(test['home'].map(nh_dict['home'])).fillna(0)

#awayでの人気のなさ
na=(pd.DataFrame(data=df_sorted_n['away'])).apply(pd.value_counts)
na_dict = na.to_dict(orient='dict')

test['not_pop_away']=(test['away'].map(na_dict['away'])).fillna(0)
```

```python
test['pop_home']=(test['home'].map(ph_dict['home'])).fillna(0)
```

```python
test.loc[:,['pop_home','pop_away','not_pop_home','not_pop_away']]
```

```python
test['home_popularity']=test['pop_home']-test['not_pop_home']
test['away_popularity']=test['pop_away']-test['not_pop_away']
```

```python
#best_eleven
best_eleven=(set(['西川\u3000周作','那須\u3000大亮','森重\u3000真人','中澤\u3000佑二','中村\u3000俊輔','山口\u3000螢','柿谷\u3000曜一朗','青山\u3000敏弘','大迫\u3000勇也','大久保\u3000嘉人','川又\u3000堅碁', ' 駒野\u3000友一',' 田中\u3000マルクス闘莉王','水本\u3000裕貴','レアンドロ\u3000ドミンゲス','遠藤\u3000保仁','青山\u3000敏弘','高萩\u3000洋次郎','ウイルソン','佐藤\u3000寿人','豊田\u3000陽平']))

h_list=[]
test['best_eleven']=1000

for i in range(len(test)):
    h_list.append(set([test['home_01'][i], test['home_02'][i],test['home_03'][i], test['home_04'][i],test['home_05'][i], test['home_06'][i], test['home_07'][i], test['home_08'][i],test['home_09'][i], test['home_10'][i], test['home_11'][i],test['away_01'][i],test['away_02'][i],test['away_03'][i],test['away_04'][i],test['away_05'][i],test['away_06'][i],test['away_07'][i],test['away_08'][i],test['away_09'][i],test['away_10'][i],test['away_11'][i]]))
for j in range(len(test)):
        test['best_eleven'][j]=len(h_list[j]&best_eleven)
```

```python
#第何節か、J1/J2に分けて集計
test['J1_section_num']=0
for i in range(len(test)):
    if test['stage'][i]=='Ｊ１':
        test['J1_section_num'][i]=int(test['match'][i].split("節")[0][1:])
```

```python
test['J2_section_num']=0
for i in range(len(test)):
    if test['stage'][i]=='Ｊ２':
        test['J2_section_num'][i]=int(test['match'][i].split("節")[0][1:])
```

```python
#J1 節ごと観客数平均を出す
ls=[0.0]*34
count=[0]*34
x1=np.arange(34)+[1.0]*34

for i in range(34):
    for j in range(len(df)):
        if df['J1_section_num'][j]==i+1:
            ls[i]+=df['y'][j]
            count[i]+=1
for k in range(len(ls)):
    ls[k]/=count[k]
```

```python
#J2 節ごと観客数平均を出す
ls2=[0.0]*42
count2=[0]*42
x2=np.arange(42)+[1.0]*42

for i in range(42):
    for j in range(len(df)):
        if df['J2_section_num'][j]==i+1:
            ls2[i]+=df['y'][j]
            count2[i]+=1
for k in range(len(ls2)):
    ls2[k]/=count2[k]
```

```python
#J1/J2節を変数化
#J1
test['J1_section_val']=0
for i in range(34):
    for j in range(len(test)):
        if test['J1_section_num'][j]==i+1:
            test['J1_section_val'][j]=ls[i]
```

```python
#J2
test['J2_section_val']=0
for i in range(42):
    for j in range(len(test)):
        if test['J2_section_num'][j]==i+1:
            test['J2_section_val'][j]=ls2[i]
```

```python
test['J1_section_val']/=1000
test['J2_section_val']/=1000
test['section_val']=test['J1_section_val']+test['J2_section_val']
```
