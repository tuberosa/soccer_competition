```python
import pandas as pd
```

```python
#データの結合
```

```python
#train data cat
train=pd.read_csv('train.csv')
train_add=pd.read_csv('train_add.csv')
train=pd.concat([train,train_add])

#condition data cat
condition=pd.read_csv('condition.csv')
condition_add=pd.read_csv('condition_add.csv')
condition=pd.concat([condition,condition_add])
```

```python
#conditionとtrain_dataを結合
train=pd.merge(train,condition,on='id')
```

```python
test=pd.read_csv('test.csv')
```

```python
#conditionとtestの結合
test=pd.merge(test,condition,on='id')
```

```python
#結合したデータの保存
train.to_csv('train_cat.csv')
test.to_csv('test_cat.csv')
```
