```python
import pandas as pd
```

```python
df=pd.read_csv('with_section.csv')
```

```python
#天気をダミー変数化する
for i in range(len(df)):
    if '雨' in df['weather'][i]:
        df['weather'][i]='rain'
    elif '雪' in df['weather'][i]:
        df['weather'][i]='rain'
    elif '曇' in df['weather'][i]:
        df['weather'][i]='cloudy'
    elif '屋内' in df['weather'][i]:
        df['weather'][i]='indoor'
    elif '晴' in df['weather'][i]:
        df['weather'][i]='sunny'
```

```python
temp3=pd.get_dummies(df['weather'])
df['rain']=temp3['rain']
df['cloudy']=temp3['cloudy']
df['indoor']=temp3['indoor']
df['sunny']=temp3['sunny']
```

```python
df.loc[:,['weather','rain','cloudy','sunny','indoor']]
```

```python
df.head()
```

```python
#曜日のダミー変数化
df['fri']=0
df['sat']=0
df['sun']=0
df['holiday']=0
df['weekday']=0



for i in range(len(df)):
    if '金' in df['gameday'][i]:
        df['fri'][i]=1
    elif '土' in df['gameday'][i]:
        df['sat'][i]=1
    elif '日' in df['gameday'][i]:
        df['sun'][i]=1
    elif '祝' in df['gameday'][i]:
        df['holiday'][i]=1
    else:
        df['weekday'][i]=1
```

```python
df.loc[:,['fri','sat','sun','holiday','weekday']]
```

```python
#NHKs:NHK総合で放映されていたら1そうでなかったら0
df['NHKs']=0
for i in range(len(df)):
    if 'ＮＨＫ総合' in df['tv'][i]:
        df['NHKs'][i]=1
```
