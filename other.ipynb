```python
#popularity
df_sorted=df.sort_values('y',ascending=True)
df_sorted_p=df_sorted.tail(250)
df_sorted_n=df_sorted.head(250)
```

```python
df_sorted.tail()
```

```python
ph=(pd.DataFrame(data=df_sorted_p['home'])).apply(pd.value_counts)
ph_dict = ph.to_dict(orient='dict')
ph_dict['home']
```

```python
df['home'].head()
```

```python
df['pop_home']=(df['home'].map(ph_dict['home'])).fillna(0)
df['pop_home']
```

```python
df['pop_home']=(df['home'].map(ph_dict)).fillna(0)
```

```python
df['pop_home']
```

```python
#away_popularity
pa=(pd.DataFrame(data=df_sorted_p['away'])).apply(pd.value_counts)
pa_dict = pa.to_dict(orient='dict')
pa_dict['away']
```

```python
df['pop_away']=(df['away'].map(pa_dict['away'])).fillna(0)
df['pop_away'].head()
```

```python
df=df.drop('away_home',axis=1)
```

```python
#homeでの人気のなさ
nh=(pd.DataFrame(data=df_sorted_n['home'])).apply(pd.value_counts)
nh_dict = nh.to_dict(orient='dict')

df['not_pop_home']=(df['home'].map(nh_dict['home'])).fillna(0)
df['not_pop_home']
```

```python
#awayでの人気のなさ
na=(pd.DataFrame(data=df_sorted_n['away'])).apply(pd.value_counts)
na_dict = na.to_dict(orient='dict')

df['not_pop_away']=(df['away'].map(na_dict['away'])).fillna(0)
df['not_pop_away']
```
